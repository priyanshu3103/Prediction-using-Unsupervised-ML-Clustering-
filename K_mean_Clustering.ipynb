```python
x = data.iloc[:, [0, 1, 2, 3]].values
```

```python
# Finding the number of clusters for K-means classification by using Elboe method

wcss = []

for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++',max_iter = 300, n_init= 10 , random_state= 0)  
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title("The Elboe method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# Apply K-means to the dataset for prediction

kmeans = KMeans(n_clusters=3, init='k-means++', random_state= 0)  
y_predict= kmeans.fit_predict(x) 
y_predict 
```

```python
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], s = 100, c = 'purple', label = 'Iris-setosa')
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], s = 100, c = 'orange', label = 'Iris-versicolour')
plt.scatter(x[y_predict == 2, 0], x[y_predict == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')

#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'red', label = 'Centroids')

plt.legend()
```
